```python
# df=df_weather.merge(df_climbing ,on ='Date') 
joined_table= pd.merge(climbing_data, weather_data, how="left", on=["Date"])
```

```python
joined_table
```

```python
# cambiar los nan de temperatura y demas por datos extraidos de la api
#incluir niveles de dificultad de las rutas
```

```python
# tehre is s strong relation between success and success rate so i drop the success columsn
joined_table_clean=joined_table.drop(labels="Succeeded", axis=1)
```

```python
set(joined_table_clean["Success Percentage"])
```

```python
# there are success percentage values higher than 1 which is not right because percentaje should be below 1 so i make all the percentejes above 1, 1
def corrected_percentage(row):
    if row["Success Percentage"]>1:
        return 1
    else:
        return row["Success Percentage"]
    
joined_table_clean["Success Percentage"]=joined_table_clean.apply(corrected_percentage, axis=1)
```

```python
set(joined_table_clean["Success Percentage"])
```

```python
# create two new columns with month and year
joined_table_clean['Year'] = pd.DatetimeIndex(joined_table_clean['Date']).year
joined_table_clean['Month'] = pd.DatetimeIndex(joined_table_clean['Date']).month
```

```python
joined_table_clean
```

```python
#  change farehit to degrees celsius
def far_cel(F):
    C=(F-32)*(5/9)
    return C

joined_table_clean["Temperature AVG"]= joined_table_clean["Temperature AVG"].apply(far_cel)
```

```python
joined_table_clean
```

```python
#change miles per hour to km per hour in wind speed
def m_km(miles):
    km= 1.60934 * miles
    return km

joined_table_clean["Wind Speed Daily AVG"]= joined_table_clean["Wind Speed Daily AVG"].apply(m_km)
joined_table_clean
```
