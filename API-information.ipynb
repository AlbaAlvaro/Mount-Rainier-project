### Obtaining the data with Onecall Option

```python
'''A more useful way of getting the data with Onecall.
Shows the weather for day 08/11/2020 at 17:00, minute fortcast for 1 hour, hourly forecast, daily forecast 7 days and the
historical weather for the 5 previous days. '''

api_key = f"{op_token}"
lat = "46.879967"
lon = "-121.726906"
url = "https://api.openweathermap.org/data/2.5/onecall?lat=%s&lon=%s&appid=%s&units=metric" % (lat, lon, api_key)
dt="1604361600"

response = requests.get(url)
data = json.loads(response.text)
print(str(data)[:500])
```

```python
# Obtain the temperatures from 8 Nov to 15 Nov 2020
temps=[]
days=[]
daily = data["daily"]

for entry in daily:
    temp = entry["temp"]
    day= entry["dt"]
    temps.append(temp)
    days.append(dt)

print(temps)
print(days)
```

```python
# Obtain the humidity from 8 Nov to 15 Nov 2020
humidity=[]
days=[]
daily = data["daily"]

for entry in daily:
    hum = entry["humidity"]
    day= entry["dt"]
    humidity.append(hum)
    days.append(dt)
print(humidity)
print(days)
```

```python
# Obtain the wind speed from 8 Nov to 15 Nov 2020
wind_s=[]
days=[]
daily = data["daily"]

for entry in daily:
    wind = entry["wind_speed"]
    day= entry["dt"]
    wind_s.append(wind)
    days.append(dt)
print(wind_s)
print(days)
```

The data for solar radiation and wind direction can not be obtained with this API

```python
# Collecting the temperature for each day from 8 Nov to 15 Nov 2020

df = pd.json_normalize(data["daily"][0]["temp"])
print (df)
df1= pd.json_normalize(data["daily"][1]["temp"])
print(df1)
df2 = pd.json_normalize(data["daily"][2]["temp"])
print (df2)
df3= pd.json_normalize(data["daily"][3]["temp"])
print(df3)
df4 = pd.json_normalize(data["daily"][4]["temp"])
print (df4)
df5= pd.json_normalize(data["daily"][5]["temp"])
print(df5)
df6 = pd.json_normalize(data["daily"][6]["temp"])
print (df6)
df7= pd.json_normalize(data["daily"][7]["temp"])
print(df7)
```

```python
# Concateneting in a dataframe all the data from temperature
result = pd.concat([df,df1, df2, df3, df4, df5, df6, df7])
result
```

```python
# Adding a Date column to specify the date of each of the measurements
result["Date"]=["08/11/2020", "09/11/2020", "10/11/2020", "11/11/2020", "12/11/2020", "13/11/2020", "14/11/2020", "15/11/2020"]
result
```

```python
# Collecting the temperature for each day from 8 Nov to 15 Nov 2020

dict_0 = {"humidity" : data["daily"][0]["humidity"], "wind_speed": data["daily"][0]["wind_speed"] }
dict_1 = {"humidity" : data["daily"][1]["humidity"], "wind_speed": data["daily"][1]["wind_speed"] }
dict_2 = {"humidity" : data["daily"][2]["humidity"], "wind_speed": data["daily"][2]["wind_speed"] }
dict_3 = {"humidity" : data["daily"][3]["humidity"], "wind_speed": data["daily"][3]["wind_speed"] }
dict_4 = {"humidity" : data["daily"][4]["humidity"], "wind_speed": data["daily"][4]["wind_speed"] }
dict_5 = {"humidity" : data["daily"][5]["humidity"], "wind_speed": data["daily"][5]["wind_speed"] }
dict_6 = {"humidity" : data["daily"][6]["humidity"], "wind_speed": data["daily"][6]["wind_speed"] }
dict_7 = {"humidity" : data["daily"][7]["humidity"], "wind_speed": data["daily"][7]["wind_speed"] }

df0hw = pd.json_normalize(dict_0)
df1hw = pd.json_normalize(dict_1)
df2hw = pd.json_normalize(dict_2)
df3hw = pd.json_normalize(dict_3)
df4hw = pd.json_normalize(dict_4)
df5hw = pd.json_normalize(dict_5)
df6hw = pd.json_normalize(dict_6)
df7hw = pd.json_normalize(dict_7)

result_h_w = pd.concat([df0hw,df1hw, df2hw, df3hw, df4hw, df5hw, df6hw, df7hw])
result_h_w
```

```python
# COncatenate the values of both tables
joined_thw = pd.concat([result, result_h_w], axis=1)
joined_thw
```

```python
# Giving an index to the rows of the table
s = pd.Series([0, 1, 2, 3, 4, 5, 6, 7])
joined_thw.set_index([s])
```

```python
# Rename the columns as in the cleaned DataFrame
joined_clean=joined_thw.rename(columns={"day": "Temperature AVG"})
joined_clean1=joined_clean.rename(columns={"humidity": "Relative Humidity AVG"})
joined_clean2=joined_clean1.rename(columns={"wind_speed": "Wind Speed Daily AVG"})

joined_clean2
```

```python
# Delete the not useful columns
joined_clean3=joined_clean2.drop(labels=["min", "max", "night", "eve", "morn"], axis=1)
joined_clean3
```

```python
# Changing from miles/hour to kilometers/hour the wind speed

def m_km(miles):
    km= 1.60934 * miles
    return km

joined_clean3["Wind Speed Daily AVG"]= joined_clean3["Wind Speed Daily AVG"].apply(m_km)
joined_clean3
```

## Join the data of the previos DataFrame from data cleaning to this new DataFrame

```python
weather_climbing_data
```

```python
# # Insert the new values in the table from data cleaning
cleaned_df=weather_climbing_data.append(joined_clean3)
cleaned_df
```

```python
# Delete column "Unnamed:0"
cleaned_df=cleaned_df.drop(labels="Unnamed: 0", axis=1)
cleaned_df
```

```python
# Add the year and month from the new entries to their respective columns
cleaned_df['Year'] = pd.DatetimeIndex(cleaned_df['Date']).year
cleaned_df['Month'] = pd.DatetimeIndex(cleaned_df['Date']).month
cleaned_df
```
